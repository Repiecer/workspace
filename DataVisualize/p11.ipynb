```python
import openpyxl
```

```python
data_target = openpyxl.load_workbook('./ALLINONE.xlsx')
mint_target = data_target.active
```

```python
year = 2023
season = 1
filename = f'{year}年第{season}季度废水废气实际排放量计算结果.xlsx'
data_source = openpyxl.load_workbook(filename)
mint_source = data_source.active
```

```python
def get_space():
    sp = 1
    while True:
        val = mint_target.cell(row=2, column=sp).value
        if val == None:
            break
        sp+=1
    return sp
```

```python
loc = [6, 10, 14]
```

```python
for i in loc:
    st = get_space()
    for j in range(12):
        mint_target.cell(row=2+j, column=st, value=mint_source.cell(row=14+j, column=i).value)
```

```python
data_target.save('./ALLINONE.xlsx')
data_target.close()
data_source.close()
```
