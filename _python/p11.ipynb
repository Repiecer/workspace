```python
import string
db = ['Bob2049', 'Alice2026']
while True:
    name:str = input('Enter username:')
    if len(name) <6:
        print('must be at least six charactors.')
        continue
    if name[0] not in string.ascii_letters:
        print('must begin with alphabets.')
        continue
    if name.isalnum() and any(c.isdigit() for c in name) and any(c.isalpha() for c in name):
        print('ok')
        db.append(name)
        break
    else:
        print('must be alphabets and numbers')
print('users:', db)
```

```python
import sys
while True:

    a, b, c = map(float, input('a b c:').split())
    if all(t==0 for t in [a,b,c]):
        break
    if a == 0:
        print('wrong a')
        continue
    delta = b**2-4*a*c
    x1 = (-b+(abs(delta)**0.5)*(1j if delta<0 else 1))/(2*a)
    x2 = (-b-(abs(delta)**0.5)*(1j if delta<0 else 1))/(2*a)
    print(f'{x1 = :.2f} {x2 = :.2f}')
```

```python
n = int(input())
def isprime(n):
    if n<=1:return False
    for i in range(2, n):
        if n%i==0:return False
    return True
if isprime(n):print('YES')
else:print('No')
```
